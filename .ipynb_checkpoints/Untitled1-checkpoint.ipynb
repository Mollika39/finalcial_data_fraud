```python
# Feature Engineering
# Example: Creating a feature for transaction amount frequency (dummy feature here for illustration)
cd['AmountFrequency'] = cd['Amount'] / (cd['Time'] + 1)  # Avoid division by zero
```

```python
# Handling Categorical Data (if any)
# cd['Categorical_Column'] = LabelEncoder().fit_transform(cd['Categorical_Column'])

# Preparing data for modeling
X = cd.drop('Class', axis=1)
Y = cd['Class']
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Feature Selection using RFE
model = LogisticRegression(max_iter=1000)  # Increase max_iter if convergence warning appears
rfe = RFE(model, n_features_to_select=10)  # Select top 10 features
fit = rfe.fit(X_scaled, Y)
print(f"Selected Features: {X.columns[fit.support_]}")
```

```python
# Dimensionality Reduction using PCA
pca = PCA(n_components=10)  # Reduce to top 10 components
X_pca = pca.fit_transform(X_scaled)
print(f"Explained variance ratio by PCA: {pca.explained_variance_ratio_}")
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
# Model Training and Evaluation
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier()
}
```

```python
for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Hyperparameter Tuning
    if name == 'Logistic Regression':
        param_grid = {'C': [0.01, 0.1, 1, 10, 100]}
        grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
        grid_search.fit(X_train, Y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model
    
    # Fit the model
    best_model.fit(X_train, Y_train)
    
    # Predictions and Evaluation
    ypred = best_model.predict(X_test)
    
    accuracy = accuracy_score(Y_test, ypred) * 100
    balanced_accuracy = balanced_accuracy_score(Y_test, ypred) * 100
    precision = precision_score(Y_test, ypred) * 100
    recall = recall_score(Y_test, ypred) * 100
    f1 = f1_score(Y_test, ypred) * 100
    kappa = cohen_kappa_score(Y_test, ypred)
    
    print(f"{name} Metrics:")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"Balanced Accuracy: {balanced_accuracy:.2f}%")
    print(f"Precision: {precision:.2f}%")
    print(f"Recall: {recall:.2f}%")
    print(f"F1 Score: {f1:.2f}%")
    print(f"Cohen's Kappa Score: {kappa:.2f}")
    
    # Cross-Validation
    cv_scores = cross_val_score(best_model, X_pca, Y, cv=5, scoring='accuracy')
    print(f"Cross-Validation Accuracy Scores: {cv_scores}")
    print(f"Mean Cross-Validation Accuracy: {cv_scores.mean() * 100:.2f}%")
    
    # Visualization: ROC Curve
    fpr, tpr, _ = roc_curve(Y_test, best_model.predict_proba(X_test)[:,1])
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'{name} ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic - {name}')
    plt.legend(loc="lower right")
    plt.show()
    
    # Visualization: Feature Importances (for Random Forest and Decision Tree)
    if name in ['Random Forest', 'Decision Tree']:
        importances = best_model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.figure()
        plt.title(f'Feature Importances - {name}')
        plt.bar(range(X_pca.shape[1]), importances[indices], align='center')
        plt.xticks(range(X_pca.shape[1]), indices, rotation=90)
        plt.xlim([-1, X_pca.shape[1]])
        plt.show()
```

```python
import streamlit as st
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load and preprocess data
data = pd.read_csv(r"C:\Users\KIIT\Desktop\credit\creditcard.csv")

# Separate legitimate and fraudulent transactions
legit = data[data.Class == 0]
fraud = data[data.Class == 1]

# Undersample legitimate transactions to balance the classes
legit_sample = legit.sample(n=len(fraud), random_state=2)
data = pd.concat([legit_sample, fraud], axis=0)

# Feature engineering
data['AmountFrequency'] = data['Amount'] / (data['Time'] + 1)  # Avoid division by zero
data['TimeSquared'] = data['Time'] ** 2
data['AmountTimeRatio'] = data['Amount'] / (data['Time'] + 1)  # Adjust based on domain knowledge

# Prepare data for modeling
X = data.drop('Class', axis=1)
y = data['Class']

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dimensionality Reduction using PCA
pca = PCA(n_components=10)  # Reduce to top 10 components
X_pca = pca.fit_transform(X_scaled)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, stratify=y, random_state=42)

# Create models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}

# Create dictionaries to store metrics
accuracy_scores = {}
roc_auc_scores = {}

# Train and evaluate models
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores[model_name] = accuracy
    
    # ROC AUC Score
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:,1])
    roc_auc = auc(fpr, tpr)
    roc_auc_scores[model_name] = roc_auc

# Streamlit app
st.title("Credit Card Fraud Detection")

st.subheader("Model Comparison")

# Display accuracy scores
st.write("Model Accuracy Scores:")
st.write(accuracy_scores)

# Display ROC AUC Scores
st.write("Model ROC AUC Scores:")
st.write(roc_auc_scores)

# Plot accuracy comparison
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(accuracy_scores.keys(), accuracy_scores.values(), color=['green', 'blue', 'red'])
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy Comparison')
plt.xticks(rotation=45)
st.pyplot(fig)

# Plot ROC Curves
plt.figure(figsize=(8, 6))
for model_name in models.keys():
    fpr, tpr, _ = roc_curve(y_test, models[model_name].predict_proba(X_test)[:,1])
    roc_auc = roc_auc_scores[model_name]
    plt.plot(fpr, tpr, label=f'{model_name} (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
st.pyplot(plt.gcf())

# Display classification reports
st.write("Classification Reports:")
for model_name, model in models.items():
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    st.write(f"Classification Report for {model_name}:")
    st.dataframe(report_df.style.background_gradient(cmap='coolwarm'))

# Predict with RandomForest model
model = RandomForestClassifier()
model.fit(X_train, y_train)

st.subheader("Predict Transaction Type")
st.write("Enter the following features to check if the transaction is legitimate or fraudulent:")

# Input fields for user to enter feature values
input_values = []
features = X.columns

for feature in features:
    value = st.number_input(f'{feature}', value=0.0)
    input_values.append(value)

# Create a button to submit input and get prediction
submit = st.button("Submit")

if submit:
    # Ensure input has the correct shape
    features = np.array(input_values).reshape(1, -1)
    # Standardize and apply PCA
    features_scaled = scaler.transform(features)
    features_pca = pca.transform(features_scaled)
    # Make prediction
    prediction = model.predict(features_pca)
    # Display result
    if prediction[0] == 0:
        st.title("Legitimate Transaction")
    else:
        st.title("Fraudulent Transaction")
```
